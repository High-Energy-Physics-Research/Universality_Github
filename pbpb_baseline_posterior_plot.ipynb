# Pb–Pb baseline posterior plot for $w$

This notebook reproduces the **baseline 1D posterior plot** for the Gaussian nucleon width $w$ using the **Pb–Pb** posterior files:

- `posterior/mcmc_chain_Grad.csv` → **Only scaled spectra calibration**
- `posterior/new_LHC_posterior_samples.csv` → **Only $p_T$-integrated calibration**

It mirrors the style of the Xe–Xe analysis notebook, but switches the input to the Pb–Pb baseline files.

> Assumption: this notebook is saved in the **repository root** (same level as `posterior/`).


```
import os
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Resolve repository root robustly.
# Preferred layout:
# repo_root/
#   posterior/
#   this_notebook.ipynb

candidate_roots = [
    Path.cwd(),
    Path.cwd().parent,
    Path('/mnt/data')
]

repo_root = None
for root in candidate_roots:
    if (root / 'posterior').exists():
        repo_root = root
        break

if repo_root is None:
    raise FileNotFoundError(
        "Could not find a 'posterior/' directory. "
        "Place this notebook in the repository root or adjust `repo_root` manually."
    )

posterior_dir = repo_root / 'posterior'
posterior_dir
```

```
# Input files for the Pb–Pb baseline
scaled_file = posterior_dir / 'mcmc_chain_Grad.csv'
integrated_file = posterior_dir / 'new_LHC_posterior_samples.csv'

print('Scaled spectra posterior:', scaled_file)
print('Integrated-observables posterior:', integrated_file)

if not scaled_file.exists():
    raise FileNotFoundError(f"Missing file: {scaled_file}")

if not integrated_file.exists():
    raise FileNotFoundError(f"Missing file: {integrated_file}")
```

```
posterior_scaled_df = pd.read_csv(scaled_file)
posterior_integrated_df = pd.read_csv(integrated_file)

print('scaled shape     =', posterior_scaled_df.shape)
print('integrated shape =', posterior_integrated_df.shape)

display(posterior_scaled_df.head())
display(posterior_integrated_df.head())
```

## Choose the parameter to plot

In the Xe–Xe notebook, `observables_to_plot = 3` corresponded to the Gaussian nucleon width $w$.

This cell keeps that as the default, but also prints the columns so you can confirm the index-name mapping in your local files.


```
for i, col in enumerate(posterior_scaled_df.columns):
    print(i, col)
```

```
param_index = 3  # expected to be w
param_name = posterior_scaled_df.columns[param_index]
param_name
```

## Baseline posterior plot

This reproduces the **same logic** as the Xe–Xe notebook, now for **Pb–Pb**:

- blue: only scaled spectra calibration
- orange: only $p_T$-integrated calibration


```
sns.set_theme(style='white')

fig_width = 8.6
fig_height = 6.45
font_size = 18

fig, ax = plt.subplots(figsize=(fig_width, fig_height))

ax.tick_params(
    axis='both',
    which='both',
    direction='in',
    bottom=True,
    top=True,
    left=True,
    right=True,
    labelsize=font_size,
    width=1.5,
    length=4,
    labelleft=True,
)

sns.kdeplot(
    data=posterior_scaled_df,
    x=param_name,
    fill=True,
    label='Only scaled spectra calibration',
    color='blue',
    linewidth=2,
    alpha=0.7,
    ax=ax,
)

sns.kdeplot(
    data=posterior_integrated_df,
    x=param_name,
    fill=True,
    label='Only $p_T$-integrated calibration',
    color='orange',
    linewidth=2,
    alpha=0.7,
    ax=ax,
)

ax.legend(loc='best', fontsize=13, frameon=True, edgecolor='black')
ax.set_xlabel(param_name, fontsize=18)
ax.set_ylabel('Posterior Distribution', fontsize=18)
ax.set_title('PARAMETER POSTERIORS: GAUSSIAN NUCLEON WIDTH', fontsize=18)

plt.tight_layout()
plt.show()
```
